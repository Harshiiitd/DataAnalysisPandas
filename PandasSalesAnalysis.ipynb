```python
# arr=[1,2,3,4,5,6,7,8,9]

df['month']=df['Order Date'].apply(lambda x:str(x)[0:2])
df['month']
```

first we need to change the dtype of price to float

```python
tr=df['Price Each'].isnull()
tr.value_counts()
```

```python
df.dropna(axis="index",how="all",subset=['Price Each'],inplace=True)
tr=df['Price Each'].isnull()
tr.value_counts()
```

```python
df['month'].value_counts()
```

*Month column has an or type so we need to remove it*

```python
df=df.loc[(df['month']!='Or')]
df['month'].value_counts()
```

*Change month column to int*

```python
df['month']=df['month'].astype('int')
```

*Change price each to float*

```python
df['Price Each']=df['Price Each'].astype('float')
```

*Calculate the total sale using sales=each_prize * quantity*

```python
df['Quantity Ordered']=df['Quantity Ordered'].astype('int')
```

```python
df['sales']=df['Quantity Ordered'] * df['Price Each']
```

```python
df.head(10)
```

```python
month_grouped=df.groupby('month')
```

```python
ma=month_grouped["sales"].sum()
ma
```

```python
x=range(1,13)
plt.bar(x,ma)
plt.xlabel('months')
plt.ylabel('sales')
plt.show()
```

#Getting product who sold together most in pair of 2 or 3 or 4

```python
from itertools import combinations
from collections import Counter
```

```python
i=0
for row in new_df['grouped']:
  if(i==4):
      break
  row_list=row.split(',')
  i=i+1
  print(row_list)
```

```python
count=Counter()

for row in new_df['grouped']:
  row_list=row.split(',')
  count.update(Counter(combinations(row_list,2)))

count.most_common(10)
```

```python
count=Counter()

for row in new_df['grouped']:
  row_list=row.split(',')
  count.update(Counter(combinations(row_list,3)))

count.most_common(10)
```

```python
count=Counter()

for row in new_df['grouped']:
  row_list=row.split(',')
  count.update(Counter(combinations(row_list,4)))

count.most_common(10)
```
